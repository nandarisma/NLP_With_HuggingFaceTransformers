```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This movie explores the complexities of human relationships and emotionsy",
    candidate_labels=["psychology", "entertainment", "technology"],
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
   "This novel delves into themes of resilience and hope during difficult times",
    candidate_labels=['literature', 'psychology', 'motivation'],
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
   "The workshop focuses on techniques for effective time management and productivity",
    candidate_labels=['self-help', 'business', 'education'],
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("Our seminar covers the latest advancements in <mask> technology.", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("Dr. Martinez joined OpenAI and is based in San Francisco.")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Matahari adalah bintang di pusat tata surya yang menghasilkan energi melalui fusi nuklir."
question = "Apa itu Matahari?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Soeharto (lahir 8 Juni 1921 - meninggal 27 Januari 2008) adalah presiden Indonesia yang menjabat dari tahun 1967 hingga 1998. Ia dikenal sebagai pemimpin yang otoriter dan memiliki pengaruh besar dalam politik Indonesia selama lebih dari tiga dekade."
question = "Siapa Soeharto?"
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("The product exceeded my expectations, and I would recommend it to everyone.")
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
  """
    The rise of artificial intelligence (AI) has significantly impacted various industries
    across the globe. Companies are increasingly integrating AI technologies into their
    operations to enhance efficiency, streamline processes, and improve decision-making.
    While AI offers numerous advantages, it also raises ethical concerns regarding
    job displacement and data privacy. As AI continues to evolve, businesses and
    policymakers must navigate these challenges to harness its potential effectively.
    Furthermore, the collaboration between AI and human intelligence is expected to drive
    innovation and create new opportunities in the future.
    """
)
```

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")

text_to_translate = "Saya senang bermain sepak bola"
result = translator(text_to_translate)
print(result[0]['translation_text'])
```


1. **Zero Shot Classification**: Zero shot classification adalah teknik untuk mengklasifikasikan teks ke dalam kategori yang belum pernah dilihat sebelumnya tanpa memerlukan data pelatihan spesifik untuk kategori tersebut. Ini dilakukan dengan memanfaatkan pemahaman umum model tentang hubungan antara kategori dan konteks teks.

2. **Text Generation**: Text generation adalah proses di mana model menghasilkan teks baru berdasarkan input yang diberikan, seperti frasa atau kalimat awal. Teknik ini sering digunakan dalam aplikasi seperti pembuatan konten otomatis, chatbot, dan penulisan kreatif.

3. **Fill Mask**: Fill mask adalah teknik yang digunakan untuk mengisi kata yang hilang dalam suatu kalimat dengan memprediksi kata yang paling sesuai berdasarkan konteks sekitarnya. Ini berguna dalam tugas pemahaman bahasa alami, di mana model harus menebak kata yang tepat dari konteks.

4. **Named Entity Recognition (NER)**: NER adalah tugas dalam pengolahan bahasa alami yang bertujuan untuk mengidentifikasi dan mengklasifikasikan entitas bernama dalam teks, seperti nama orang, lokasi, atau organisasi. Ini membantu dalam ekstraksi informasi penting dan analisis data.

5. **Question Answering**: Question answering adalah sistem yang dirancang untuk memberikan jawaban langsung terhadap pertanyaan yang diajukan dalam bahasa alami. Ini melibatkan pemahaman konteks dan informasi yang relevan dari teks yang lebih besar atau basis data.

6. **Sentiment Analysis**: Sentiment analysis adalah proses menentukan sentimen atau opini yang terkandung dalam teks, apakah positif, negatif, atau netral. Ini sering digunakan dalam analisis ulasan produk, umpan balik pelanggan, dan media sosial.

7. **Summarization**: Summarization adalah teknik untuk menghasilkan ringkasan dari teks yang lebih panjang, menyajikan informasi inti tanpa kehilangan makna penting. Ini berguna dalam membantu pengguna untuk cepat memahami isi dokumen atau artikel.

8. **Translation**: Translation adalah proses mengubah teks dari satu bahasa ke bahasa lain sambil mempertahankan makna asli. Dengan kemajuan dalam pemrosesan bahasa alami, model kini mampu melakukan terjemahan yang lebih akurat dan alami.
